```
import requests   #importing requests and BeautifulSoup for web scraping
from bs4 import BeautifulSoup as bs
url = "https://www.worldometers.info/coronavirus/"    #URL of coronavirus informations site
response = requests.get(url)    #Getting content or information from webpage
# soup = bs(data,"html.parser")
# print(soup)
# data.headers['content-type']
# print(response.text)
# print(response.content)
```

```
soup = bs(response.content,'html5lib')   #Parsing HTML content
# sp = soup.prettify()
```

```
data = soup.find('table', attrs = {'class':'table table-bordered table-hover main_table_countries'})   #Find table of corona virus status in html form
head_details = []   #Heading column of corona data
# cell_details = []
row_details = []    #create empty lists. rows of corona details country wise

for heading in data.findAll('th'):
  head_details.append(heading.getText())

for row in data.findAll('tr'):
  cell_details = []
  for cell in row.findAll('td'):
    cell_details.append(cell.getText())
  row_details.append(cell_details)
row_details = row_details[1:]   #All corona status in list form
# print(head_details)
# for i in row_details:
#   print(i)

#len(row_details)

all_datas = []  #getting all information in Dictionary form into a single list. Like Json file data
row_count = 0
for i in row_details:
  all_data = {}
  col_count = 0
  for j in head_details:
    all_data[j] = row_details[row_count][col_count]
    col_count += 1
  all_datas.append(all_data)
  row_count += 1

# print(all_datas)  #print all corona virus cases country wise
```
