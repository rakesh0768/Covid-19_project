```
import folium
import json
import import_ipynb
import covid19_cases_data as ccd

path = 'countries.json'
with open(path,'r') as f:
  countries_location = json.load(f)

m = folium.Map(tiles = "Stamen Toner", min_zoom = 2, width = '80%', max_zoom=4)
fg = folium.FeatureGroup(name = "corona")


#Define popup values for countries
def corona_values(country_name):
  x = "data not found"
  for i in ccd.all_datas:
    if str(country_name) in i['Country,Other']:
      x=str(country_name) + "\nCases=" + str(i['TotalCases']) + "\nDeaths=" + str(i['TotalDeaths']) + "\nActive=" + str(i['ActiveCases'])
      break
  return x

#definr circle redius acording to
def cir(country_name):
  r = 0
  for i in ccd.all_datas:
    if str(country_name) in i['Country,Other']:
      if 0<int(i['TotalCases'].replace(',',''))<=10:
        r = 5
      elif 10<int(i['TotalCases'].replace(',',''))<=1000:
        r = 10
      elif 1000<int(i['TotalCases'].replace(',',''))<=10000:
        r = 15
      elif 10000<int(i['TotalCases'].replace(',',''))<=100000:
        r = 20
      elif 100000<int(i['TotalCases'].replace(',',''))<=200000:
        r = 25
      else :
        r = 30
      break
  return r


# Test purpose
# mark = fg.add_child(folium.Marker(location=[20,78],popup='T.C.=120dkfhkdjfghsdkhgdskfhglsdfhglsdfhld0 T.D.-25'))
for point in countries_location:
  mark = fg.add_child(folium.CircleMarker(location=point['latlng'], radius=cir(str(point['name'])), color=None, popup=corona_values(str(point['name'])), fill_color='red', fill_opacity = cir(str(point['name']))/30))
  m.add_child(mark)
m.add_child(fg)
# m.add_child(folium.LayerControl())
# print(ccd.all_datas)
```
